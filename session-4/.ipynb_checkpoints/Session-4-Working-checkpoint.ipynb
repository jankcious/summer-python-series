## PART 5 – Slightly more advanced stuff

The scouts and analysts want to create (slightly) more advanced pitching stats from existing data so they can better evaluate the players’ performances.

Create a new dataframe with calculations of : SO/BB, SO/9, P/Inning, PAB, WHIP for each pitcher from data in the stacked dataframe (not the filtered dataframes). For those not familiar with these baseball stat terms, here is a quick breakdown:

- SO/BB: strikeouts per walk ratio  
- SO/9: strikeouts per 9 innings pitched  
- P/Inning: pitches per inning  
- PAB: pitches per at-bat ratio  
- WHIP: (BB + H) per inning pitched  

The .groupby() and .sum() functions could be of great use here!!! 

It may be advantageous to review the columns of the stacked/master dataframe created in PART 3

Caution/Hint: The way baseball innings are recorded is quite terrible modular arithmetic. For example, “4.1” innings is actually 4 and 1/3 innings. “4.2” innings is actually 4 and 2/3 innings. Therefore, “4.1” + “4.2” innings is not “8.3” innings, but actually 9 innings. There might need to be a new column created in order to add innings correctly…
  
This new data frame should have columns in this order and no other columns: 
['Pitcher', 'IP', 'H', 'SO', 'BB', 'WHIP', 'SO/BB', 'SO/9', 'H/9', 'P/Inning', 'PAB']

Export this new dataframe as a .csv file named "2015 Houston Astros extra stats.csv" without an index column


```python
Master_Original = Master
```

```python
stats = pd.DataFrame(columns = ['Pitcher', 'IP', 'H', 'SO', 'BB', 'WHIP', 'SO/BB', 'SO/9', 'H/9', 'P/Inning', 'PAB'])
stats = pd.DataFrame(columns = ['Pitcher'])
```

```python
Master.columns
```

```python
stats['Pitcher'] = Master['Pitcher'].unique()
```

```python
Master['IP'] = Master['IP'].astype(float) // 1 + (Master['IP'].astype(float) % 1 * 10 / 3)
Master['IP'].values
```

```python
IP = pd.DataFrame(Master.groupby(by = 'Pitcher')['IP'].mean())
stats = pd.merge(left = stats, right = IP, left_on = 'Pitcher', right_index = True)
```

```python
H = pd.DataFrame(Master.groupby(by = 'Pitcher')['H'].mean())
stats = pd.merge(left = stats, right = H, left_on = 'Pitcher', right_index = True)
```

```python
SO = pd.DataFrame(Master.groupby(by = 'Pitcher')['SO'].mean())
stats = pd.merge(left = stats, right = SO, left_on = 'Pitcher', right_index = True)
```

```python
BB = pd.DataFrame(Master.groupby(by = 'Pitcher')['BB'].mean())
stats = pd.merge(left = stats, right = BB, left_on = 'Pitcher', right_index = True)
```

```python
stats['WHIP'] = (stats['BB'] + stats['H']) / stats['IP']
```

```python
stats['SO/BB'] = stats['SO'] / stats['BB']
```

```python
Master['SO/9'] = Master['SO'] / (Master['Innings'] / 9)
SO9 = pd.DataFrame(Master.groupby(by = 'Pitcher')['SO/9'].mean())
stats = pd.merge(left = stats, right = SO9, left_on = 'Pitcher', right_index = True)
```

```python
Master['H/9'] = Master['H'] / (Master['Innings'] / 9)
H9 = pd.DataFrame(Master.groupby(by = 'Pitcher')['H/9'].mean())
stats = pd.merge(left = stats, right = H9, left_on = 'Pitcher', right_index = True)
```

```python
Master['P/Inning'] = Master['Pit'] / Master['Innings']
```

```python
PInning = pd.DataFrame(Master.groupby(by = 'Pitcher')['P/Inning'].mean())
stats = pd.merge(left = stats, right = PInning, left_on = 'Pitcher', right_index = True)
```

```python
Master['PAB'] = Master['Pit'] / Master['AB']
PAB = pd.DataFrame(Master.groupby(by = 'Pitcher')['PAB'].mean())
stats = pd.merge(left = stats, right = PAB, left_on = 'Pitcher', right_index = True)
```
